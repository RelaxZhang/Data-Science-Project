# Group the region with highly positive correlation

```python
import pandas as pd

cor_mat  = pd.read_csv("../Data/totpop1000_corrmatrix.csv")
name_area = list(cor_mat.columns)[1:]

group_list = []
already_store = []

for i in name_area:
    group_area = []
    if i in already_store:
        continue
    else:
        group_area.append(i)
        already_store.append(i)
        for j in range(len(cor_mat[i])):
            if i != name_area[j]:
                if cor_mat[i][j] >= 0.95:
                    if name_area[j] in already_store:
                        continue
                    else:
                        group_area.append(name_area[j])
                        already_store.append(name_area[j])
        group_list.append(group_area)

for i in group_list:
    print(i)
    print("\n")
```
