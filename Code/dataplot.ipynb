# Plot of data visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pop_data = pd.read_csv("../Data/above1000.csv")

Age = ['0-4','5-9','10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64','65-69','70-74','75-79','80-84','85+']
male = pop_data.iloc[3121][11:29].tolist()
raw_female = pop_data.iloc[3121][29:47].tolist()
female = []
for i in raw_female:
    female.append(-i)

dataframe_ts = pd.DataFrame(Age)
dataframe_ts.columns = ["Age"]
dataframe_ts["Male"] = male
dataframe_ts["Female"] = female

# Plot pyramid plots for visulising data and observe data sparsity of specific hierachical group
ax1 = sns.barplot(x='Female', y='Age', data=dataframe_ts, order=Age, color = "#000000")
ax2 = sns.barplot(x='Male', y='Age', data=dataframe_ts, order=Age, color = "#42b7bd")
plt.title("Population")
plt.xlabel("Gladstone - Biloela 2000 -- Female / Male")
plt.grid()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pop_ts_data = pd.read_csv("../Data/ts_1000.csv")
popts_2000 = pop_ts_data.iloc[0:12026]["2000"]

range_start = 0
range_end = 37
avg_list = []

for i in range(325):
    avg_list.append(pop_ts_data.iloc[range_start:range_end]["2000"].tolist())
    range_start += 37
    range_end += 37

Age_sex = ['m0-4','m5-9','m10-14','m15-19','m20-24','m25-29','m30-34','m35-39','m40-44','m45-49','m50-54','m55-59','m60-64','m65-69','m70-74','m75-79','m80-84','m85+', 'f0-4','f5-9','f10-14','f15-19','f20-24','f25-29','f30-34','f35-39','f40-44','f45-49','f50-54','f55-59','f60-64','f65-69','f70-74','f75-79','f80-84','f85+', 'total']
age_count_dict = {}
for i in range(325):
    for j in range(len(Age_sex)):
        if Age_sex[j] not in age_count_dict:
            age_count_dict[Age_sex[j]] = avg_list[i][j]
        else:
            age_count_dict[Age_sex[j]] += avg_list[i][j]

age_count_avg_dict = {}
for i in age_count_dict:
    age_count_avg_dict[i] = age_count_dict[i] / 325

pyramid_2000_avg = []
for i in age_count_avg_dict:
    if (i != "total"):
        if ("f" in i):
            pyramid_2000_avg.append(-age_count_avg_dict[i])
        else:
            pyramid_2000_avg.append(age_count_avg_dict[i])

dataframe_ts = pd.DataFrame(Age)
dataframe_ts.columns = ["Age"]
dataframe_ts["Male"] = pyramid_2000_avg[0:18]
dataframe_ts["Female"] = pyramid_2000_avg[18:37]

# Plot pyramid plots for visulising data and observe data sparsity of specific hierachical group
ax1 = sns.barplot(x='Female', y='Age', data=dataframe_ts, order=Age, color = "#000000")
ax2 = sns.barplot(x='Male', y='Age', data=dataframe_ts, order=Age, color = "#42b7bd")
plt.title("Population")
plt.xlabel("325 Regions' Average Population 2000 -- Female / Male")
plt.grid()
```
