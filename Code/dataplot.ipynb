# Plot of data visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pop_data = pd.read_csv("../Data/above1000.csv")

Age = ['0-4','5-9','10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-64','65-69','70-74','75-79','80-84','85+']
male = pop_data.iloc[3373][11:29].tolist()
raw_female = pop_data.iloc[3373][29:47].tolist()
female = []
for i in raw_female:
    female.append(-i)

dataframe_ts = pd.DataFrame(Age)
dataframe_ts.columns = ["Age"]
dataframe_ts["Male"] = male
dataframe_ts["Female"] = female
```

```python
# Plot pyramid plots for visulising data and observe data sparsity of specific hierachical group
ax1 = sns.barplot(x='Female', y='Age', data=dataframe_ts, order=Age, color = "#000000")
ax2 = sns.barplot(x='Male', y='Age', data=dataframe_ts, order=Age, color = "#42b7bd")
plt.title("Population")
plt.xlabel("Female / Male")
plt.grid()
```
