```python
def area_data(data, name):
    df_area = data[data['SA3 Name'] == name]
    df_area.reset_index()
    df_area.drop(['index','S/T Code', 'S/T Name','GCCSA Code', 'GCCSA Name','SA4 Code','SA4 Name', 'SA3 Code', 'SA3 Name'], axis=1, inplace=True)
    df_area = df_area.T
    column_name = df_area.loc['Year', :]
    df_area.columns = column_name
    df_area.drop(['Year'], axis = 0, inplace = True)
    return df_area
```
